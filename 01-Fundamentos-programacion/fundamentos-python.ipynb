## Funciones

```python
def solo_saludar():
  print("Hola, cómo estás?")

solo_saludar()


def saludar(nombre, edad=None):
    """
    Todo lo que yo escriba es un comentario
    Incluidas las lineas siguentes
    """
    """Función con parámetro opcional"""
    if edad:
        return f"Hola {nombre}, tienes {edad} años"
    return f"Hola {nombre}"

def saludar_flujo(nombre, edad=None):
  # verificar si la edad NO ES None
  if edad is not None:
    return f"Hola {nombre}, tienes {edad} años"
  # Si no tiene edad devolver el nombre solamente
  else:
    return f"Hola {nombre}"

print(saludar_flujo("Fernando", 35))
print(saludar("Fernando"))
print(saludar("Fernando", 35))

def saludar_grupo_edad(nombre, edad=None):
    if edad:
        if edad >= 18:
            return f"Hola {nombre}, tienes {edad} años y eres mayor de edad"
        else:
            return f"Hola {nombre}, tienes {edad} años y eres menor de edad"
    return f"Hola {nombre}"

print(saludar_grupo_edad("Fernando", 35))
```

```python
def clasificar_edad_persona(nombre, edad):
  if edad >= 0 and edad < 18:
    return f"La persona {nombre} tiene {edad} años y es menor de edad"
  elif edad >= 18 and edad < 65:
    return f"La persona {nombre} tiene {edad} años y es un adulto"
  elif edad >= 65:
    return f"La persona {nombre} tiene {edad} años y es un anciano"
  else:
    return f"La persona {nombre} no tienen una edad definida"

print(clasificar_edad_persona("Fernando", 35))
print(clasificar_edad_persona("Fernando", -1))
```

# Estructuras de Datos

```python
# Listas
# Un arreglo de valores por posición 0, 1, 2 ... n
# Para crear listas usamos los []
edades_curso = [17, 17, 22, 17, 17, 17, 35]
print(f"Edades del curso : {edades_curso}")

# Clase o tipo de la lista
print(f"El objeto edades_curso es de tipo : {type(edades_curso)}")

# Atributos de una lista : numero de elementos
n_valores = len(edades_curso)
print(f"La lista edades_curso tiene {n_valores} valores")
# Atributos de la lista : acceder a valores por posicion
print(f"La primera edad de la lista edades_curso es: {edades_curso[0]} años")
print(f"La última edad de la lista edades_curso es : {edades_curso[-1]} años")
```

```python
# Operaciones con listas
# Segmentar una lista [inicio:fin]
edades_2 = edades_curso[0:3]
print(f"Lista edades 1 hasta 3 : {edades_2}")
# Segementar desde la posición 2 hasta el 4
edades_3 = edades_curso[1:4]
print(f"Lista edades posición 2 - 4 : {edades_3}")
# Segementar hasta la posición 3 
edades_4 = edades_curso[:3]
print(f"Lista edades de 1-3 : {edades_4}")
# Segmentar desde 1 hasta 2 método corto
print(f"Lista edades 1-2 : {edades_curso[:2]}")
# Segmentar los ultimos tres elementos
print(f"Lista tres últimas edades : {edades_curso[-3:]}")
```

```python
# Operaciones sobre elementos de la lista
# Agregar un nuevo valor a lista
edades_curso = [17, 17, 22, 17, 17, 17, 35]
edades_curso.append(45)
print(f"Edades actualizadas : {edades_curso}")
# Insertar nuevo elemento en una posicion
edades_curso_2 = [23, 56, 78]
for edad in edades_curso_2:
    edades_curso.append(edad)
print(f"Edades completas : {edades_curso}")
```

```python
# Eliminar elementos de una lista
edades_curso.remove(78)
print(f"Eliminar edad 78 de la lista {edades_curso}")
```

```python
# Eliminar por posicion
edades_curso.pop(2) # elemento numero 3
print(f"Eliminar el elemento en la posición 3 (index 2) {edades_curso}")
```

```python
# Insertar elementos en una lista
edades_curso.insert(4, 50)
print(f"Insertar edad 50 en la posición 5 (index 4) {edades_curso}")
```

```python
# Contar el numero de elementos que coinciden con un valor
n_edades_17 = edades_curso.count(17)
print(f"Personas con edad 17 : {n_edades_17}")
```

## Diccionarios

Los diccionarios guardan pares de clave-valor.

```python
# Crear un diccionario
persona = {
    "nombre": "Ana",
    "edad": 25,
    "ciudad": "Madrid",
    "profesion": "Ingeniera"
}

# Acceder a valores
print("Nombre:", persona["nombre"])
print("Edad:", persona["edad"])
print("Ciudad:", persona["ciudad"])
print(f"Profesion: {persona.get('profesion')}")

```

```python
# Diccionario y listas
edades_curso = {
    "nombre_curso" : "Programación CDAT",
    "nombre_alumnos" : ["Wilson", "Santiago", "Leo"],
    "edades" : [17, 17, 17]
}
# Edades del curso
print(edades_curso["edades"])

# Imprimir mensaje usando los datos del diccionario
for clave, valor in edades_curso.items():
    print(f"{clave} : {valor}")

list_iter = range(len(edades_curso["nombre_alumnos"]))
for i in list_iter:
    alumnos = edades_curso.get("nombre_alumnos")
    edades = edades_curso.get("edades")
    print(f"El estudiante {alumnos[i]} tiene {edades[i]} años")
```

```python
list_iter = range(len(edades_curso["nombre_alumnos"]))
for i in list_iter:
    alumnos = edades_curso.get("nombre_alumnos")
    edades = edades_curso.get("edades")
    nombre_profesor = edades_curso.get("nombre_profesor")
    print(f"El estudiante {alumnos[i]} tiene {edades[i]} años y está en la clase de profesor {nombre_profesor}")
```

# Manejo de Excepciones

```python
# Dividir por cero arroja un error
def dividir_error(a, b):
    resultado = a / b
    return f"Resultado: {resultado}"

print(dividir_error(10, 2))
print(dividir_error(10, 0))
```

```python
# division segura
def dividir_seguro(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        return "Error: No se puede dividir por cero"
    except TypeError:
        return "Error: Tipo de dato incorrecto"
    else:
        return f"Resultado: {resultado}"
    finally:
        print("Operación finalizada")

print(dividir_seguro(10, 2))
print(dividir_seguro(10, 0))
print(dividir_seguro("hola", 0))
```

## Pandas: Análisis de Datos

```python
import pandas as pd

# Crear DataFrame
datos = {
    'Nombre': ['Ana', 'Luis', 'María', 'Carlos', 'Elena'],
    'Edad': [28, 32, 25, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Madrid', 'Sevilla'],
    'Salario': [45000, 52000, 41000, 58000, 47000]
}

df = pd.DataFrame(datos)

print("DataFrame:")
print(df)
print(f"\nEstadísticas:")
print(df.describe())
```

## Pandas: Operaciones

```python
# Filtrado
madrid = df[df['Ciudad'] == 'Madrid']
print("Empleados en Madrid:")
print(madrid)

# Agrupación
por_ciudad = df.groupby('Ciudad')['Salario'].mean()
print("\nSalario promedio por ciudad:")
print(por_ciudad)

# Ordenamiento
ordenado = df.sort_values('Salario', ascending=False)
print("\nOrdenado por salario:")
print(ordenado[['Nombre', 'Salario']])
```

## Matplotlib: Visualización

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear gráfico
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Grafico de funciones trigonometricas
ax1.plot(x, y1, 'b-', label='Seno')
ax1.plot(x, y2, 'r--', label='Coseno')
ax1.set_title('Funciones Trigonométricas')
ax1.legend()
ax1.grid(True)

# Gráfico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]
ax2.bar(categorias, valores, color=['red', 'blue', 'green', 'orange'])
ax2.set_title('Gráfico de Barras')
ax2.set_ylabel('Valores')

plt.tight_layout()
plt.show()
```
