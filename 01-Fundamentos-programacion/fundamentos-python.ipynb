## Funciones

```python
def saludar(nombre, edad=None):
    """Función con parámetro opcional"""
    if edad:
        return f"Hola {nombre}, tienes {edad} años"
    return f"Hola {nombre}"

print(saludar("Ana", 30))
print(saludar("Carlos", 20))
```

# Estructuras de Datos

```python
# Creación y manipulación de listas
frutas = ["manzana", "banana", "cereza"]
numeros = [1, 2, 3, 4, 5]

# Operaciones comunes
frutas.append("naranja")
frutas.insert(0, "uva")
frutas.pop(1)
print(f"Frutas: {frutas}")

# Slicing
print(f"Primeras 3 frutas: {frutas[:3]}")
print(f"Últimas 2 frutas: {frutas[-2:]}")
```

## List Comprehensions

```python
# Crear listas de forma elegante
cuadrados = [x**2 for x in range(10)]
print(f"Cuadrados: {cuadrados}")

# Con condición
pares = [x for x in range(20) if x % 2 == 0]
print(f"Números pares: {pares}")

# Transformación
nombres = ["ana", "juan", "maría"]
mayusculas = [nombre.upper() for nombre in nombres]
print(f"Mayúsculas: {mayusculas}")
```

## Diccionarios

```python
# Diccionarios: pares clave-valor
persona = {
    "nombre": "Carlos",
    "edad": 35,
    "ciudad": "Madrid",
    "profesion": "Ingeniero",
    "Direccion" : "Cll 34-45"
}

print(f"Nombre: {persona['nombre']}")
print(f"Edad: {persona.get('edad')}")

# Agregar, modificar y borrar
persona["email"] = "carlos@example.com"
persona["edad"] = 36
persona.pop("Direccion")

print(f"\nDiccionario actualizado:")
# iterar clave valor
for clave, valor in persona.items():
    print(f"  {clave}: {valor}")
```

## Sets (Conjuntos)

```python
# Sets: colecciones sin duplicados
numeros = {1, 2, 3, 4, 5}
pares = {2, 4, 6, 8}

# Operaciones de conjuntos
union = numeros | pares
interseccion = numeros & pares
diferencia = numeros - pares

print(f"Unión: {union}")
print(f"Intersección: {interseccion}")
print(f"Diferencia: {diferencia}")
```

```python
# Tuplas: inmutables
coordenadas = (10, 20)

# Desempaquetado
x, y = coordenadas
print(f"X: {x}, Y: {y}")

# No se pueden modificar (nos arroja un error)
coordenadas[0] = 1
```

# Manejo de Excepciones

```python
# Dividir por cero arroja un error
def dividir_error(a, b):
    resultado = a / b
    return f"Resultado: {resultado}"

print(dividir_error(10, 2))
print(dividir_error(10, 0))
```

```python
# division segura
def dividir_seguro(a, b):
    try:
        resultado = a / b
    except ZeroDivisionError:
        return "Error: No se puede dividir por cero"
    except TypeError:
        return "Error: Tipo de dato incorrecto"
    else:
        return f"Resultado: {resultado}"
    finally:
        print("Operación finalizada")

print(dividir_seguro(10, 2))
print(dividir_seguro(10, 0))
print(dividir_seguro("hola", 0))
```

## Pandas: Análisis de Datos

```python
import pandas as pd

# Crear DataFrame
datos = {
    'Nombre': ['Ana', 'Luis', 'María', 'Carlos', 'Elena'],
    'Edad': [28, 32, 25, 35, 29],
    'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Madrid', 'Sevilla'],
    'Salario': [45000, 52000, 41000, 58000, 47000]
}

df = pd.DataFrame(datos)

print("DataFrame:")
print(df)
print(f"\nEstadísticas:")
print(df.describe())
```

## Pandas: Operaciones

```python
# Filtrado
madrid = df[df['Ciudad'] == 'Madrid']
print("Empleados en Madrid:")
print(madrid)

# Agrupación
por_ciudad = df.groupby('Ciudad')['Salario'].mean()
print("\nSalario promedio por ciudad:")
print(por_ciudad)

# Ordenamiento
ordenado = df.sort_values('Salario', ascending=False)
print("\nOrdenado por salario:")
print(ordenado[['Nombre', 'Salario']])
```

## Pandas: Leer y exportar datos

```python
# Leer un archivo de texto
df_desde_csv = pd.read_csv("datos.csv")
print("\nLeer desde un CSV:")
print(df_desde_csv)

# Exportar a un archivo de texto
df_filtrado = df[df['Ciudad'] == 'Madrid']
print("\nDatos a exportar a CSV:")
print(df_filtrado)
df_filtrado.to_csv("datos_madrid.csv", index=False)
# Leer
df_madrid = pd.read_csv("datos_madrid.csv")
print("\nLeer datos nuevos exportados:")
print(df_madrid)
```

## Matplotlib: Visualización

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Crear gráfico
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Grafico de funciones trigonometricas
ax1.plot(x, y1, 'b-', label='Seno')
ax1.plot(x, y2, 'r--', label='Coseno')
ax1.set_title('Funciones Trigonométricas')
ax1.legend()
ax1.grid(True)

# Gráfico de barras
categorias = ['A', 'B', 'C', 'D']
valores = [23, 45, 56, 78]
ax2.bar(categorias, valores, color=['red', 'blue', 'green', 'orange'])
ax2.set_title('Gráfico de Barras')
ax2.set_ylabel('Valores')

plt.tight_layout()
plt.show()
```
