# Manejo de Errores

```r
# Dividir insegura
dividir_no_seguro <- function(a, b) {
  resultado <- a / b
  return(resultado)
}

cat("División válida:", dividir_no_seguro(10, 2), "\n")
cat("División por cero:", dividir_no_seguro(10, 0), "\n")
```

```r
# tryCatch para manejo de errores
dividir_seguro <- function(a, b) {
  tryCatch(
    {
      # Intento de dividir
      resultado <- a / b
      return(resultado)
    },
    warning = function(w) {
      message("Advertencia: ", w)
      return(NA)
    },
    error = function(e) {
      message("Error: No se puede dividir entre cero.")
      return(NA)
    },
    finally = {
      message("Operación finalizada.")
    }
  )
}

cat("División válida:", dividir_seguro(10, 2), "\n")
cat("División por cero:", dividir_seguro(10, 0), "\n")
cat("División con valores inválidos:", dividir_seguro("Hola", 2), "\n")
```

## dplyr: Manipulación de Datos

```r
library(dplyr)

# Crear datos de ejemplo
empleados <- data.frame(
  Nombre = c("Ana", "Luis", "María", "Carlos", "Elena", "Pedro"),
  Edad = c(28, 32, 25, 35, 29, 31),
  Departamento = c("IT", "Ventas", "IT", "Ventas", "IT", "RRHH"),
  Salario = c(45000, 52000, 41000, 58000, 47000, 44000)
)

# Pipeline con %>%
resultado <- empleados %>%
  filter(Departamento == "IT") %>%
  select(Nombre, Salario) %>%
  arrange(desc(Salario))

print("Empleados de IT ordenados por salario:")
print(resultado)
```

## dplyr: Transformaciones

```r
# Operaciones avanzadas con dplyr
resumen <- empleados %>%
  group_by(Departamento) %>%
  summarise(
    Empleados = n(),
    Salario_Promedio = mean(Salario),
    Salario_Max = max(Salario),
    .groups = 'drop'
  ) %>%
  arrange(desc(Salario_Promedio))

print("Resumen por departamento:")
print(resumen)
```

## dplyr: Mutate y Transformaciones

```r
# Crear nuevas columnas
empleados_trans <- empleados %>%
  mutate(
    Salario_Anual = Salario * 12,
    Categoria = case_when(
      Salario > 50000 ~ "Senior",
      Salario > 45000 ~ "Mid",
      TRUE ~ "Junior"
    ),
    Experiencia_Estimada = (Edad - 22)
  ) %>%
  select(Nombre, Categoria, Salario_Anual, Experiencia_Estimada)

print("Empleados con categorías:")
print(empleados_trans)
```

## tidyr: Reestructuración de Datos


```r
library(tidyr)

# Datos anchos
ventas_wide <- data.frame(
  Producto = c("Laptop", "Mouse", "Teclado"),
  Enero = c(10, 50, 30),
  Febrero = c(15, 45, 35),
  Marzo = c(12, 55, 32)
)

# Convertir a formato largo
ventas_long <- ventas_wide %>%
  pivot_longer(
    cols = c(Enero, Febrero, Marzo),
    names_to = "Mes",
    values_to = "Ventas"
  )

print("Formato largo:")
print(ventas_long)
```

## tidyr: Pivot Wider

```r
# Convertir de vuelta a formato ancho
ventas_wide_nuevo <- ventas_long %>%
  pivot_wider(
    names_from = Mes,
    values_from = Ventas
  )

print("De vuelta a formato ancho:")
print(ventas_wide_nuevo)

# Separar columnas
datos_split <- data.frame(
  Nombre_Completo = c("Ana García", "Luis Martínez", "María López")
) %>%
  separate(Nombre_Completo, into = c("Nombre", "Apellido"), sep = " ")

print("Nombres separados:")
print(datos_split)
```

## ggplot2: Visualización de Datos

```r
library(ggplot2)

# Gráfico de dispersión
p1 <- ggplot(empleados, aes(x = Edad, y = Salario)) +
  geom_point(aes(color = Departamento), size = 4) +
  geom_smooth(method = "lm", se = FALSE) +
  labs(title = "Salario vs Edad por Departamento",
       x = "Edad (años)",
       y = "Salario ($)") +
  theme_minimal() +
  theme(legend.position = "bottom")

print(p1)
```

## ggplot2: Gráficos de Barras


```r
# Gráfico de barras con dplyr
resumen_dept <- empleados %>%
  group_by(Departamento) %>%
  summarise(Salario_Promedio = mean(Salario), .groups = 'drop')

p2 <- ggplot(resumen_dept, aes(x = reorder(Departamento, Salario_Promedio), 
                                y = Salario_Promedio, fill = Departamento)) +
  geom_col() +
  geom_text(aes(label = paste0("$", round(Salario_Promedio/1000, 1), "k")), 
            vjust = -0.5) +
  labs(title = "Salario Promedio por Departamento",
       x = "Departamento",
       y = "Salario Promedio ($)") +
  theme_minimal() +
  theme(legend.position = "none")

print(p2)
```
