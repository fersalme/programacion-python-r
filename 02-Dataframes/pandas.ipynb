## 1. 🐼 ¿Qué es Pandas?

**Pandas** es la biblioteca más importante de Python para **manipulación y análisis de datos tabulares**.

### ¿Por qué Pandas?

| Herramienta | Límite | Velocidad | Facilidad |
|-------------|--------|-----------|-----------|
| Excel | ~1M filas | Lenta | Alta |
| Python listas | Sin límite | Media | Media |
| NumPy arrays | Sin límite | Muy alta | Baja (sin etiquetas) |
| **Pandas** | **Sin límite** | **Alta** | **Muy alta** |

### Dos estructuras principales

```
pandas
├── Series    → array 1D etiquetado  (como una columna de Excel)
└── DataFrame → tabla 2D etiquetada  (como una hoja de Excel completa)
```

> 💡 Pandas está construido **sobre NumPy**: internamente usa `ndarray`, pero añade etiquetas (índices y nombres de columnas) que hacen el trabajo con datos mucho más intuitivo.

## 2. 📊 Series — Array 1D etiquetado

Una **Serie** es el bloque básico de Pandas: un array de una dimensión con un **índice** asociado a cada valor.

```
Índice │ Valor
───────┼───────
  0    │  120
  1    │  340
  2    │  210
  3    │  480
```

````python
import pandas as pd
import numpy as np

# ── Creación de Series ──────────────────────────────────────────

# Desde lista (índice automático 0, 1, 2...)
ventas = pd.Series([120, 340, 210, 480, 300])
print("Serie básica:")
print(ventas)

# Desde lista con índice personalizado
temperaturas = pd.Series(
    [22.5, 19.3, 25.1, 28.7, 21.0],
    index=["Lun", "Mar", "Mié", "Jue", "Vie"],
    name="Temperatura (°C)"
)
print("Serie con índice personalizado:")
print(temperaturas)
````

````python
import pandas as pd

# Desde diccionario (clave → índice, valor → dato)
poblacion = pd.Series({
    "Bogotá":    7743955,
    "Medellín":  2533424,
    "Cali":      2227642,
    "Barranquilla": 1228621,
    "Cartagena":  1028736
}, name="Población 2023")

print(poblacion)
print()
print("dtype:    ", poblacion.dtype)
print("shape:    ", poblacion.shape)
print("size:     ", poblacion.size)
print("name:     ", poblacion.name)
````

### 2.1 Acceso e indexación en Series

````python
import pandas as pd

s = pd.Series([10, 20, 30, 40, 50],
              index=["a", "b", "c", "d", "e"])
print("Serie s:", s)

# Por etiqueta (label-based)
print("s['a']     =", s["a"])        # 10
print("s[['a','c']] =", s[["a","c"]])

# Por posición (position-based)
print("s.iloc[0]  =", s.iloc[0])     # 10
print("s.iloc[-1] =", s.iloc[-1])    # 50
print("s.iloc[1:4] =", s.iloc[1:4])

# Indexación booleana
print("s[s > 25]  =", s[s > 25])
````

### 2.2 Operaciones con Series

````python
import pandas as pd
import numpy as np

a = pd.Series([1, 2, 3, 4], index=["w", "x", "y", "z"])
b = pd.Series([10, 20, 30, 40], index=["w", "x", "y", "z"])

# Aritmética — alineación automática por índice
print("a + b:", a + b)
print("a * 100:", a * 100)

# Alineación con índices distintos → NaN donde no coincide
c = pd.Series([100, 200, 300], index=["x", "y", "z"])
print("a + c (índices distintos):", a + c)

# Funciones estadísticas
ventas = pd.Series([120, 340, 210, 480, 300, 150, 520])
print("Estadísticas de ventas:")
print(ventas.describe())
````

## 5. 🔍 Selección e indexación

Esta es una de las habilidades más importantes en Pandas. Hay tres métodos principales:

| Método | Tipo de acceso | Sintaxis |
|--------|---------------|---------|
| `[]` | Columnas o filas | `df["col"]`, `df[0:3]` |
| `.loc[]` | Por **etiqueta** | `df.loc[idx, "col"]` |
| `.iloc[]` | Por **posición** entera | `df.iloc[0, 2]` |

````python
import pandas as pd

df = pd.DataFrame({
    "nombre":  ["Ana", "Carlos", "Luis", "María", "Pedro"],
    "edad":    [28, 35, 22, 31, 45],
    "ciudad":  ["Bogotá", "Medellín", "Cali", "Bogotá", "Barranquilla"],
    "salario": [4_500_000, 6_200_000, 2_800_000, 5_100_000, 7_300_000]
})

print("DataFrame original:")
print(df)
print()

# ── Selección de columnas ─────────────────────────────────────
print("Una columna (Series):")
print(df["nombre"])
print()
print("Varias columnas (DataFrame):")
print(df[["nombre", "salario"]])
````

````python
import pandas as pd

df = pd.DataFrame({
    "nombre":  ["Ana", "Carlos", "Luis", "María", "Pedro"],
    "edad":    [28, 35, 22, 31, 45],
    "ciudad":  ["Bogotá", "Medellín", "Cali", "Bogotá", "Barranquilla"],
    "salario": [4_500_000, 6_200_000, 2_800_000, 5_100_000, 7_300_000]
})

# ── .loc[] — por ETIQUETA ──────────────────────────────────────
print("df.loc[0] — fila con índice 0:")
print(df.loc[0])
print()

print("df.loc[1:3] — filas índice 1 a 3 (¡incluye el 3!):")
print(df.loc[1:3])
print()

print("df.loc[0, 'nombre'] — celda específica:")
print(df.loc[0, "nombre"])
print()

print("df.loc[[0,2,4], ['nombre','salario']] — filas y columnas específicas:")
print(df.loc[[0, 2, 4], ["nombre", "salario"]])
````

````python
import pandas as pd

df = pd.DataFrame({
    "nombre":  ["Ana", "Carlos", "Luis", "María", "Pedro"],
    "edad":    [28, 35, 22, 31, 45],
    "ciudad":  ["Bogotá", "Medellín", "Cali", "Bogotá", "Barranquilla"],
    "salario": [4_500_000, 6_200_000, 2_800_000, 5_100_000, 7_300_000]
})

# ── .iloc[] — por POSICIÓN entera ─────────────────────────────
print("df.iloc[0] — primera fila (posición 0):")
print(df.iloc[0])
print()

print("df.iloc[0:3] — primeras 3 filas (excluye posición 3):")
print(df.iloc[0:3])
print()

print("df.iloc[:, 0:2] — todas las filas, primeras 2 columnas:")
print(df.iloc[:, 0:2])
print()

print("df.iloc[1, 3] — fila posición 1, columna posición 3:")
print(df.iloc[1, 3])
````

### 5.1 Filtros booleanos

````python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 20
df = pd.DataFrame({
    "nombre":    [f"Empleado_{i}" for i in range(n)],
    "edad":      np.random.randint(22, 55, n),
    "salario":   np.random.randint(2_000_000, 9_000_000, n),
    "ciudad":    np.random.choice(["Bogotá","Medellín","Cali"], n),
    "educacion": np.random.choice(["Técnico","Universitario","Posgrado"], n)
})

# Filtro simple
mayores_35 = df[df["edad"] > 35]
print("Empleados mayores de 35 años:")
print(mayores_35[["nombre","edad","salario"]].head())
print()

# Filtro compuesto con & (AND) y | (OR)
filtro = (df["salario"] > 5_000_000) & (df["ciudad"] == "Bogotá")
print("Salario > 5M Y ciudad = Bogotá:")
print(df[filtro][["nombre","salario","ciudad"]])
````

````python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 20
df = pd.DataFrame({
    "nombre":    [f"Empleado_{i}" for i in range(n)],
    "edad":      np.random.randint(22, 55, n),
    "salario":   np.random.randint(2_000_000, 9_000_000, n),
    "ciudad":    np.random.choice(["Bogotá","Medellín","Cali"], n),
    "educacion": np.random.choice(["Técnico","Universitario","Posgrado"], n)
})

# isin() — filtrar por lista de valores
ciudades_principales = ["Bogotá", "Medellín"]
print("Empleados en ciudades principales:")
print(df[df["ciudad"].isin(ciudades_principales)][["nombre","ciudad"]].head(6))
print()

# between() — rango de valores
print("Empleados con salario entre 4M y 7M:")
print(df[df["salario"].between(4_000_000, 7_000_000)][["nombre","salario"]].head(6))
print()

# str.contains() — filtro en texto
print("Empleados cuyo nombre contiene '_1':")
print(df[df["nombre"].str.contains("_1")][["nombre"]])
````

````python
import pandas as pd
import numpy as np

df_sucio = pd.DataFrame({
    "nombre":  ["Ana", "Carlos", None, "María", "Pedro", "Ana", "Luis"],
    "edad":    [28.0, 35.0, 22.0, None, 45.0, 28.0, -5.0],
    "salario": [4_500_000, 6_200_000, 2_800_000, 5_100_000, None, 4_500_000, 3_000_000],
    "ciudad":  ["Bogotá", "medellín", "CALI", "Bogotá", "Barranquilla", "Bogotá", "cali"],
})

# ── Estrategias para manejar nulos ──────────────────────────────

# 1. Eliminar filas con algún nulo
df_sin_nulos = df_sucio.dropna()
print("dropna() — eliminar filas con nulos:")
print(df_sin_nulos)
print()

# 2. Eliminar solo si TODAS las columnas son nulas
df_all_null = df_sucio.dropna(how="all")

# 3. Rellenar con un valor fijo
df_fill = df_sucio.fillna({"nombre": "Desconocido", "salario": 0})
print("fillna() — rellenar nulos:")
print(df_fill)
````

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "edad":    [28.0, 35.0, np.nan, 31.0, 45.0, np.nan, 22.0],
    "salario": [4.5e6, 6.2e6, 2.8e6, np.nan, 7.3e6, 5.1e6, np.nan],
})

# 4. Rellenar con estadísticas (la más común en ML)
df_media = df.copy()
df_media["edad"].fillna(df_media["edad"].mean(), inplace=True)
df_media["salario"].fillna(df_media["salario"].median(), inplace=True)
print("Relleno con media/mediana:")
print(df_media.round(0))
print()

# 5. Propagación hacia adelante/atrás
df_ffill = df.fillna(method="ffill")   # Forward fill
df_bfill = df.fillna(method="bfill")   # Backward fill
print("Forward fill (ffill):")
print(df_ffill)
````

### 6.2 Duplicados y tipos de datos

````python
import pandas as pd

df = pd.DataFrame({
    "nombre":  ["Ana", "Carlos", "Luis", "María", "Pedro", "Ana", "Luis"],
    "edad":    [28, 35, 22, 31, 45, 28, 22],
    "salario": [4_500_000, 6_200_000, 2_800_000, 5_100_000, 7_300_000, 4_500_000, 2_800_000],
    "ciudad":  ["Bogotá", "medellín", "CALI", "Bogotá", "Barranquilla", "Bogotá", "cali"],
})

# ── Duplicados ────────────────────────────────────────────────────
print("Filas duplicadas:")
print(df.duplicated())
print()
print(f"Total duplicados: {df.duplicated().sum()}")
print()

# Eliminar duplicados
df_sin_dup = df.drop_duplicates()
print(f"Filas originales: {len(df)}  →  Sin duplicados: {len(df_sin_dup)}")
print()

# ── Corrección de texto ───────────────────────────────────────────
df["ciudad_limpia"] = df["ciudad"].str.lower().str.capitalize()
print("Ciudad original vs limpia:")
print(df[["ciudad", "ciudad_limpia"]])
````

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "edad_str":     ["28", "35", "22", "31", "45"],
    "salario_str":  ["4500000", "6200000", "2800000", "5100000", "7300000"],
    "fecha_str":    ["2024-01-15", "2024-02-20", "2024-03-10", "2024-04-05", "2024-05-18"],
    "activo_str":   ["True", "False", "True", "True", "False"]
})

print("Tipos originales (todo string):")
print(df.dtypes)
print()

# ── Conversión de tipos ──────────────────────────────────────────
df["edad"]    = df["edad_str"].astype(int)
df["salario"] = df["salario_str"].astype(float)
df["fecha"]   = pd.to_datetime(df["fecha_str"])
df["activo"]  = df["activo_str"].map({"True": True, "False": False})

print("Tipos convertidos:")
print(df[["edad","salario","fecha","activo"]].dtypes)
print()
print(df[["edad","salario","fecha","activo"]])
````

## 7. ⚙️ Transformación de datos

Transformar datos es crear nuevas columnas, modificar las existentes o cambiar la estructura del DataFrame.

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df = pd.DataFrame({
    "nombre":    [f"Empleado_{i}" for i in range(8)],
    "salario":   [4_500_000, 6_200_000, 2_800_000, 5_100_000,
                  7_300_000, 3_200_000, 8_100_000, 4_800_000],
    "horas_sem": [40, 45, 38, 42, 50, 35, 48, 40],
    "ciudad":    ["Bogotá","Medellín","Cali","Bogotá","Barranquilla","Cali","Bogotá","Medellín"]
})

# ── Crear nuevas columnas ─────────────────────────────────────────
df["salario_hora"]   = (df["salario"] / (df["horas_sem"] * 4)).round(0)
df["salario_anual"]  = df["salario"] * 12
df["sobre_promedio"] = df["salario"] > df["salario"].mean()

print("DataFrame con columnas nuevas:")
print(df[["nombre","salario","salario_hora","salario_anual","sobre_promedio"]])
````

### 7.1 apply() — la función más flexible

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "salario": [4_500_000, 6_200_000, 2_800_000, 5_100_000, 7_300_000],
    "ciudad":  ["Bogotá", "Medellín", "Cali", "Bogotá", "Barranquilla"],
    "edad":    [28, 35, 22, 31, 45]
})

# apply() sobre columna — aplica función a cada elemento
def clasificar_salario(s):
    if s < 3_000_000:   return "Bajo"
    elif s < 6_000_000: return "Medio"
    else:               return "Alto"

df["nivel_salario"] = df["salario"].apply(clasificar_salario)
print("Con apply (función custom):")
print(df)
print()

# apply() con lambda — más conciso
df["salario_miles"] = df["salario"].apply(lambda x: f"${x/1_000_000:.1f}M")
print("Salario formateado:")
print(df[["salario","salario_miles","nivel_salario"]])
````

````python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    "nombre": ["  Ana García  ", "carlos LÓPEZ", "LUIS Martínez", "María pérez"],
    "salario": [4_500_000, 6_200_000, 2_800_000, 5_100_000],
    "ciudad":  ["Bogotá", "medellín", "CALI", "bogotá"]
})

# ── Operaciones con strings (.str) ──────────────────────────────
df["nombre_limpio"]  = df["nombre"].str.strip().str.title()
df["ciudad_limpia"]  = df["ciudad"].str.capitalize()
df["nombre_largo"]   = df["nombre_limpio"].str.len()
df["iniciales"]      = df["nombre_limpio"].str.split().apply(lambda x: ".".join([p[0] for p in x]) + ".")

print("Transformaciones de texto:")
print(df[["nombre","nombre_limpio","ciudad","ciudad_limpia","iniciales"]])
````

## 8. 📦 Agrupación y agregación con `groupby`

`groupby` es una de las funciones más poderosas de Pandas. Permite aplicar la lógica **"Split → Apply → Combine"**:

```
DataFrame original
       ↓ split
Grupos (por categoría, ciudad, etc.)
       ↓ apply
Función (sum, mean, count, custom...)
       ↓ combine
Resultado agregado
```

| Operación | Equivalente SQL | Descripción |
|-----------|----------------|-------------|
| `pd.merge()` | `JOIN` | Une por columna(s) clave |

| `df.join()` | `JOIN` por índice | Une por el índice |

````python
import pandas as pd

# Dataset de empleados
empleados = pd.DataFrame({
    "emp_id":   [1, 2, 3, 4, 5],
    "nombre":   ["Ana", "Carlos", "Luis", "María", "Pedro"],
    "dept_id":  [10, 20, 10, 30, 20]
})

# Dataset de departamentos
departamentos = pd.DataFrame({
    "dept_id":  [10, 20, 30, 40],
    "dept_nombre": ["Ingeniería", "Ventas", "RRHH", "Marketing"],
    "presupuesto": [500_000_000, 300_000_000, 150_000_000, 200_000_000]
})

print("Empleados:")
print(empleados)
print("\nDepartamentos:")
print(departamentos)
````

## 10. 🔬 Análisis Exploratorio de Datos (EDA)

El **EDA** es el proceso de examinar y resumir un dataset para entender su estructura, detectar anomalías y encontrar patrones. Pandas es la herramienta principal para esta tarea.

````python
import pandas as pd
import numpy as np

# Crear dataset realista de ventas de e-commerce
np.random.seed(42)
n = 500

categorias = ["Electrónica", "Ropa", "Hogar", "Deportes", "Libros"]
ciudades   = ["Bogotá", "Medellín", "Cali", "Barranquilla", "Cartagena"]

df = pd.DataFrame({
    "orden_id":   range(1, n+1),
    "categoria":  np.random.choice(categorias, n, p=[0.3, 0.25, 0.2, 0.15, 0.1]),
    "ciudad":     np.random.choice(ciudades, n, p=[0.4, 0.25, 0.2, 0.1, 0.05]),
    "precio":     np.random.lognormal(mean=13, sigma=1, size=n).round(0),
    "unidades":   np.random.randint(1, 10, n),
    "calificacion": np.random.choice([1,2,3,4,5], n, p=[0.05,0.1,0.2,0.35,0.3]),
    "devuelto":   np.random.choice([True, False], n, p=[0.08, 0.92])
})
df["total"] = (df["precio"] * df["unidades"]).round(0)

print("Dataset de e-commerce:")
df.info()
````

````python
# Estadísticas descriptivas completas
print("═" * 55)
print("RESUMEN ESTADÍSTICO DEL DATASET")
print("═" * 55)

print(f"\n📦 Total de órdenes: {len(df):,}")
print(f"💰 Ventas totales:   ${df['total'].sum():,.0f}")
print(f"📊 Ticket promedio:  ${df['total'].mean():,.0f}")
print(f"📈 Ticket máximo:    ${df['total'].max():,.0f}")
print(f"📉 Ticket mínimo:    ${df['total'].min():,.0f}")
print(f"↩️  Tasa devolución:  {df['devuelto'].mean()*100:.1f}%")
print(f"⭐ Calificación prom:{df['calificacion'].mean():.2f}/5")

print("\n── Ventas por categoría ────────────────────")
cat_stats = df.groupby("categoria").agg(
    ordenes=("orden_id","count"),
    total_ventas=("total","sum"),
    ticket_prom=("total","mean"),
    calif_prom=("calificacion","mean")
).sort_values("total_ventas", ascending=False)
cat_stats["total_ventas"] = cat_stats["total_ventas"].map("${:,.0f}".format)
cat_stats["ticket_prom"]  = cat_stats["ticket_prom"].map("${:,.0f}".format)
cat_stats["calif_prom"]   = cat_stats["calif_prom"].round(2)
print(cat_stats)
````

````python
# ── Distribución y correlaciones ─────────────────────────────────
print("── Distribución de calificaciones ──────────")
cal_dist = df["calificacion"].value_counts().sort_index()
for cal, cnt in cal_dist.items():
    bar = "█" * (cnt // 5)
    print(f"  {'⭐'*cal} ({cal}): {bar} {cnt}")

print("\n── Top 5 ciudades por ventas ────────────────")
ciudad_ventas = df.groupby("ciudad")["total"].sum().sort_values(ascending=False)
for ciudad, total in ciudad_ventas.items():
    print(f"  {ciudad:<15} ${total:>15,.0f}")

print("\n── Tasa de devolución por categoría ─────────")
dev_cat = df.groupby("categoria")["devuelto"].mean().sort_values(ascending=False)
for cat, tasa in dev_cat.items():
    print(f"  {cat:<15} {tasa*100:.1f}%")
````

````python
# ── Correlaciones numéricas ──────────────────────────────────────
print("── Matriz de correlación (numéricas) ───────")
numericas = df[["precio","unidades","calificacion","total"]].corr().round(3)
print(numericas)
print()

# ── value_counts() — distribución de categóricas ─────────────────
print("── Distribución de categorías ───────────────")
print(df["categoria"].value_counts())
print()
print(df["categoria"].value_counts(normalize=True).round(3) * 100)  # porcentajes
````

## 11. 📅 Series de Tiempo (Time Series)

Pandas tiene soporte nativo y muy potente para datos temporales: indexación por fechas, resampleo, ventanas deslizantes y más.

````python
import pandas as pd
import numpy as np

# ── Crear fechas con pd.date_range ──────────────────────────────
fechas_diarias = pd.date_range(start="2024-01-01", end="2024-03-31", freq="D")
print("Primeras 5 fechas diarias:")
print(fechas_diarias[:5])
print()

fechas_mensuales = pd.date_range(start="2024-01-01", periods=12, freq="MS")
print("Primeros meses del año:")
print(fechas_mensuales)
print()

# ── Crear serie temporal de ventas diarias ──────────────────────
np.random.seed(42)
n_dias = len(fechas_diarias)

ventas_ts = pd.Series(
    data  = (np.random.normal(2_000_000, 400_000, n_dias)
             + np.linspace(0, 500_000, n_dias)          # tendencia
             + 300_000 * np.sin(np.linspace(0, 4*np.pi, n_dias))  # estacionalidad
            ).clip(500_000).round(0),
    index = fechas_diarias,
    name  = "Ventas diarias"
)

print("Serie temporal de ventas (primeras 7 entradas):")
print(ventas_ts.head(7))
````

````python
import pandas as pd
import numpy as np

fechas = pd.date_range("2024-01-01", "2024-03-31", freq="D")
np.random.seed(42)
ventas = pd.Series(
    (np.random.normal(2e6, 4e5, len(fechas)) + np.linspace(0, 5e5, len(fechas))).clip(5e5).round(0),
    index=fechas, name="Ventas"
)

# ── Indexación temporal ──────────────────────────────────────────
print("Ventas de enero:")
print(ventas["2024-01"].describe().round(0))
print()

print("Ventas semana del 15 al 21 de enero:")
print(ventas["2024-01-15":"2024-01-21"])
print()

# ── Resampleo — agregar a diferente frecuencia ──────────────────
ventas_semana = ventas.resample("W").sum()
ventas_mes    = ventas.resample("ME").agg(["sum","mean","min","max"])

print("Ventas semanales:")
print(ventas_semana.round(0))
````

````python
import pandas as pd
import numpy as np

fechas = pd.date_range("2024-01-01", "2024-03-31", freq="D")
np.random.seed(42)
ventas = pd.Series(
    (np.random.normal(2e6, 4e5, len(fechas)) + np.linspace(0, 5e5, len(fechas))).clip(5e5).round(0),
    index=fechas, name="Ventas"
)

# ── Componentes de fecha ──────────────────────────────────────────
df_ts = ventas.reset_index()
df_ts.columns = ["fecha", "ventas"]
df_ts["año"]      = df_ts["fecha"].dt.year
df_ts["mes"]      = df_ts["fecha"].dt.month
df_ts["dia_sem"]  = df_ts["fecha"].dt.day_name()
df_ts["semana"]   = df_ts["fecha"].dt.isocalendar().week
df_ts["es_finde"] = df_ts["fecha"].dt.dayofweek >= 5

print("Ventas por día de la semana:")
print(df_ts.groupby("dia_sem")["ventas"].mean().round(0).sort_values(ascending=False))
print()

# ── Media móvil (rolling) ─────────────────────────────────────────
ventas_df = ventas.to_frame()
ventas_df["media_movil_7d"]  = ventas.rolling(window=7).mean()
ventas_df["media_movil_30d"] = ventas.rolling(window=30).mean()
print("Con medias móviles:")
print(ventas_df.tail(7).round(0))
````

````python
import pandas as pd
import numpy as np

# (Regenerar dataset)
np.random.seed(123); n = 100
productos = ["Laptop","Tablet","Smartphone","Monitor","Teclado","Mouse","Auriculares","Webcam","SSD","RAM"]
categorias = {"Laptop":"Computadores","Tablet":"Móviles","Smartphone":"Móviles","Monitor":"Periféricos",
              "Teclado":"Periféricos","Mouse":"Periféricos","Auriculares":"Audio","Webcam":"Periféricos",
              "SSD":"Almacenamiento","RAM":"Almacenamiento"}
prod_list = np.random.choice(productos, n)
df = pd.DataFrame({
    "fecha":pd.date_range("2024-01-01",periods=n,freq="3D"),
    "producto":prod_list,"categoria":[categorias[p] for p in prod_list],
    "precio":np.random.choice([899_000,1_200_000,2_500_000,3_800_000,85_000,75_000,150_000,200_000,280_000,120_000],n),
    "unidades":np.random.randint(1,20,n),
    "vendedor":np.random.choice(["Vendedor_A","Vendedor_B","Vendedor_C","Vendedor_D"],n),
    "descuento":np.random.choice([0,0.05,0.10,0.15,0.20],n,p=[0.5,0.2,0.15,0.1,0.05])
})
df.loc[np.random.choice(n,8,replace=False),"descuento"] = np.nan
df["total_bruto"] = df["precio"] * df["unidades"]
df["total_neto"]  = (df["total_bruto"] * (1 - df["descuento"].fillna(0))).round(0)

# ══ SOLUCIÓN 1 — Limpieza ════════════════════════════════════════
print("═"*50)
print("SOLUCIÓN 1 — Limpieza de datos")
print("═"*50)
print("\n1. Nulos por columna:")
print(df.isnull().sum()[df.isnull().sum() > 0])

df["descuento"] = df["descuento"].fillna(0)
df["tiene_descuento"] = df["descuento"] > 0
pct = df["tiene_descuento"].mean() * 100
print(f"\n2. Nulos rellenados ✓")
print(f"3. Columna 'tiene_descuento' creada ✓")
print(f"4. Porcentaje de ventas con descuento: {pct:.1f}%")
````

````python
# ══ SOLUCIÓN 2 — Selección y filtrado ═══════════════════════════
print("═"*50)
print("SOLUCIÓN 2 — Filtrado y selección")
print("═"*50)

filtro = (df["categoria"] == "Computadores") & (df["unidades"] > 5)
resultado = (df[filtro]
             [["producto","unidades","total_neto"]]
             .sort_values("total_neto", ascending=False)
             .head(5)
             .reset_index(drop=True))
print("\nTop 5 ventas de Computadores (>5 unidades):")
print(resultado)
````

````python
# ══ SOLUCIÓN 3 — Agrupación ═════════════════════════════════════
print("═"*50)
print("SOLUCIÓN 3 — Agrupación y estadísticas")
print("═"*50)

print("\n1. Total ventas netas por categoría:")
print(df.groupby("categoria")["total_neto"].sum().sort_values(ascending=False).map("${:,.0f}".format))

print("\n2. Ticket promedio por vendedor:")
print(df.groupby("vendedor")["total_neto"].mean().sort_values(ascending=False).map("${:,.0f}".format))

print("\n3. Producto más vendido (por unidades):")
top_prod = df.groupby("producto")["unidades"].sum().sort_values(ascending=False)
print(f"   → {top_prod.index[0]} con {top_prod.iloc[0]} unidades")

print("\n4. Mes con más ventas:")
df["mes_num"] = df["fecha"].dt.month
ventas_mes = df.groupby("mes_num")["total_neto"].sum().sort_values(ascending=False)
meses = {1:"Enero",2:"Febrero",3:"Marzo",4:"Abril",5:"Mayo",6:"Junio",
         7:"Julio",8:"Agosto",9:"Septiembre",10:"Octubre",11:"Noviembre",12:"Diciembre"}
print(f"   → {meses[ventas_mes.index[0]]} con ${ventas_mes.iloc[0]:,.0f}")
````

````python
# ══ SOLUCIÓN 4 — Transformación ═════════════════════════════════
print("═"*50)
print("SOLUCIÓN 4 — Transformación")
print("═"*50)

meses_abrev = {1:"Ene",2:"Feb",3:"Mar",4:"Abr",5:"May",6:"Jun",
               7:"Jul",8:"Ago",9:"Sep",10:"Oct",11:"Nov",12:"Dic"}
df["mes"] = df["fecha"].dt.month.map(meses_abrev)

def rango_precio(p):
    if p < 200_000:    return "Económico"
    elif p < 1_000_000: return "Medio"
    else:               return "Premium"

df["rango_precio"] = df["precio"].apply(rango_precio)

print("\n1 y 2. Columnas 'mes' y 'rango_precio' creadas ✓")
print("\n3. Productos por rango de precio:")
print(df["rango_precio"].value_counts())
````

````python
# ══ SOLUCIÓN 5 — Reporte completo ════════════════════════════════
print("═"*55)
print("SOLUCIÓN 5 — REPORTE RESUMEN")
print("═"*55)

print(f"\n📦 Total de órdenes:          {len(df):,}")
print(f"💰 Ventas totales brutas:      ${df['total_bruto'].sum():>15,.0f}")
print(f"💳 Ventas totales netas:       ${df['total_neto'].sum():>15,.0f}")
desc_positivos = df[df["descuento"] > 0]["descuento"]
print(f"🏷️  Descuento promedio:         {desc_positivos.mean()*100:.1f}%")

print("\n── Top 3 vendedores ──────────────────")
top_vend = df.groupby("vendedor")["total_neto"].sum().sort_values(ascending=False).head(3)
for i, (vend, total) in enumerate(top_vend.items(), 1):
    print(f"  {i}. {vend:<15} ${total:,.0f}")

print("\n── Producto con mayor descuento promedio ──")
prod_desc = df[df["descuento"]>0].groupby("producto")["descuento"].mean().sort_values(ascending=False)
print(f"  → {prod_desc.index[0]} con {prod_desc.iloc[0]*100:.1f}% descuento promedio")
````
